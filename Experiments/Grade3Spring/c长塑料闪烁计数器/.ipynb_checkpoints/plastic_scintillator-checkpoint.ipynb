```python
# -*- coding: utf-8 -*-
from physicsexp.mainfunc import *
from physicsexp.gendocx import *
```

```python
x = np.array([10*i for i in range(2,12)])
volt = np.array([4.22, 3.70, 3.26, 2.96, 2.66, 2.44, 2.20, 2.01, 1.92, 1.72])
```

```python
simple_plot(x, volt, save='1.png', 
    xlab='距离(cm)', ylab='甄别阈值(V)', title='甄别阈值随距离变化曲线')
```

```python
# lnV = [math.log(i) for i in volt]
lnV = np.log(volt)
```

```python
slope, intercept = simple_linear_plot(x, lnV, 
    save='2.png', xlab='距离(cm)', ylab='甄别阈值对数', title='甄别阈值对数随距离变化曲线')
```

```python
slope, intercept
```
