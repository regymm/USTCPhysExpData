```python
# voltage
volt = 60 * np.array([
    30, 31, 32, 33, 34, 35, 36, 37, 38, 38.5, 39, 39.5, 40, 40.5, 41, 41.5, 42, 42.5, 43, 43.5, 44, 44.5, 45
])
# scintillator count
count1 = np.array([
    44359, 43766, 44125, 44437, 44256, 44014, 44275, 44491, 44429, 44441, 44164, 44142, 44476, 44239, 
    44700, 44221, 44049, 44193, 44247, 43998, 44818, 43779, 44398
])
# scintillator background
background = 1093 / (300/60)
# mwpc count
count2 = np.array([
    6769, 7556, 8920, 11715, 17673, 23622, 33963, 41303, 43771, 44057, 43709, 43636, 43971, 43687, 
    43861, 43218, 42742, 42312, 41988, 41344, 40707, 39598, 38919
])
```

```python
# detection efficiency
eff = count2 / (count1 - background) * 100
simple_plot(volt, eff, save='1.png', xlab='高压（V）', ylab='探测效率（%）', title='多丝正比室探测效率曲线')
print('max efficiency: ', max(eff), 'when voltage is ', volt[eff.tolist().index(max(eff))], 'V')
```
