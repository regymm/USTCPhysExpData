```python
angle_th = np.array([i for i in range(20, 140)])
Egamma = 661.661e3
energy_th = np.array([Egamma / (1 + Egamma/0.511e6 * (1-math.cos(i * math.pi / 180))) for i in angle_th])
```

```python
Np = np.array([4990, 4146, 3555, 3131, 2700, 2587, 2477, 2401, 2534, 2572])
R = np.array([.4725, .5126, .5628, .6134, .6640, .7149, 0.7541, .7927, .8197, .8846])
aita = np.array([7.09, 7.54, 7.92, 8.31, 8.71, 9.10, 9.40, 9.70, 9.91, 10.10])
```

```python
a = Egamma/0.511e6
angle_th = np.array([i for i in range(30, 120)])
cross_th = np.array([2.818e-14/2 * (1+math.cos(i*math.pi/180)**2)/(1 + a*(1-math.cos(i*math.pi/180)))**2 * (1+a**2*math.cos(i*math.pi/180)**2 / ((1+math.cos(i*math.pi/180)**2) * (1 + a*(1-math.cos(i*math.pi/180))))) for i in angle_th])
```

```python
simple_plot(angle, Np/(R*aita) / (Np[0]/(R[0]*aita[0])), dot='*', show=0)
simple_plot(angle_th, cross_th/cross_th[0], dot='', lab='理论', title='相对截面', xlab='角度 度', ylab='相对截面')
```
